```python
import matplotlib.pyplot as plt

import numpy as ps

x=ps.array([0,1,2,3])

y=ps.array([3,8,1,10])

plt.plot(x,y,colors='r',ls='dotted',marker='o')

plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as ps

y=ps.array([120,90,90,60])

x=('Python','Chemistry','English','Maths')

z=(2,0,0,0)

colors=('blue','orange','green','red')

plt.pie(y,labels=x,explode=z,colors=colors)

plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as ps

y=ps.array([120,90,90,60])
colors=('blue','orange','green','red')

l=('CSE','Mech','Civil','ECE')

plt.pie(y,labels=l,colors=colors)

plt.legend(title='Subjects')

plt.show()
```
